#Download Diagnostic slides

```python
json_tamplete_path = '/content/drive/Shareddrives/Thesis/Utils/Documents/WSIS_POST_request.json'
metadata_path = '/content/drive/Shareddrives/Thesis/DataSet/TCGA-BRCA/Meta-data/TCGA_BRCA_TNBC_metadata.csv'
save_path = '/content/drive/Shareddrives/Thesis/DataSet/TCGA-BRCA/Diagnostic_WSIS/'

metadata = pd.read_csv(metadata_path)
case_IDs = metadata['Patient ID']

head = {"Content-Type": "application/json; charset=utf-8"}

text = ''
file = open("download_slides.sh", "w+")

for i in range(len(case_IDs)):
#for i in range(5):

  case_ID = case_IDs[i]

  f = open(json_tamplete_path)
  body_request = dict(json.load(f))
  body_request['filters']['content'][0]['content']['value'] = [case_ID]
  body_request['filters']['content'][1]['content']['field'] = 'files.experimental_strategy'
  body_request['filters']['content'][1]['content']['value'] = 'Diagnostic Slide'
  f.close()

  response = dict((requests.post('https://api.gdc.cancer.gov/files', json = body_request, headers=head)).json())
  if len(response['data']['hits'])!=0:
    file_id = response['data']['hits'][0]['id']

    download_sentense = f'wget -c https://api.gdc.cancer.gov/data/{file_id} -O {save_path + case_ID + ".svs"} \n'

    text = text + download_sentense
  
  sys.stdout.write(f'\rPrepare {i} images of {len(case_IDs)}')
  sys.stdout.flush()

file.write(text)
file.close()
```
